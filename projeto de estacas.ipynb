# Projeto de estacas escavadas

```python
import pandas as pd

from estacas import *
```

## Entrada do usuário

```python
tipo_estaca = 'Franki'
diametro = 0.4
```

## Capacidade de carga

```python
df = pd.read_excel('./sondagem2.xlsx')
df['tipo estaca'] = tipo_estaca.lower()
df['area estaca (m2)'] = df.apply(area_estaca, axis=1)
df['perimetro estaca (m)'] = df.apply(perimetro_estaca, axis=1)
df['k'] = df.apply(k_solo, axis=1)
df['alpha'] = df.apply(alpha_solo, axis=1)
df['f_1'] = df.apply(f1, axis=1)
df['f_2'] = 2 * df['f_1']
df['r_p (kPa)'] = df.apply(tensao_ponta, axis=1)
df['P_p (kN)'] = df.apply(carga_ponta, axis=1)
df['r_l (kPa)'] = df.apply(tensao_lateral, axis=1)
# Tensão lateral acumulada
soma_resultados = []
r_l = df['r_l (kPa)'].to_list()
for i in range(len(df)):
    if i == 0:
        soma = df.iloc[i]['r_l (kPa)']
    else:
        soma = sum(r_l[:i+1])
    soma_resultados.append(soma)
df['r_l acumulado (kPa)'] = soma_resultados
df['P_l (kN)'] = df.apply(carga_lateral, axis=1)
df['r_rd [Aoki-Veloso] (kPa)'] = df['r_l acumulado (kPa)'] + df['r_p (kPa)']
df['P_rd [Aoki-Veloso] (kN)'] = df['P_p (kN)'] + df['P_l (kN)']
df
```
